```python
menu = """
            Menu

[1] Deposito
[2] Saque
[3] Extrato
[4] Sair

"""
extrato  = ""
limite = 500
saldo = 0
numero_saques = 0
limite_saques = 3
while True:
    opcao = input(menu)
    if opcao == "1":
        valor = float(input("Digite o valor: "))
        if valor > 0:
            saldo += valor
            extrato += f"Depósito: R${valor:.2f}\n"
            print(f"Saldo: R${saldo:.2f}")
        else:
            print("Operação não realizada. Por favor, digite novamente.")

    elif opcao == "2":
        valor = float(input("Informe o valor: "))
        if valor > saldo:
            print("Operação não realizada. saldo excedido")
        elif valor > limite:
            print("Operação não realizada. Limite de valor Excedido")
        elif numero_saques >= limite_saques:
            print("Operção não concluída. Limite de saques Excedido")
        elif valor > 0:
            saldo -= valor
            extrato += f"Saque: R${valor:.2f}\n"
            print(f"Saldo: R${saldo:.2f}")
            numero_saques += 1
        else:
            print("Operação não concluida. Valor inválido.")

    elif opcao == "3":
        print("\n========== EXTRATO ==========")
        if extrato == "":
          print("Ainda não foram realizadas movimentações")
        else:
          print(extrato)
        print(f"\nSaldo: R${saldo:.2f}")

    elif opcao == "4":
        print("""========= Operação Finalizada. =========""")
        break

    else:
        print("""Operação inválida. Por favor, selecione a opção desejada.""")
```
